```python
import pandas as pd

def featurize(df):
    # Drop rows with missing values (NaN) in the 'Comment ID #' column


    X = df[['Comment ID #', 'Expected Score', 'Predicted Score']]

    # All of these have only two options (Fat Shaming and Skinny Shaming) or (Toxic and Not Toxic)
    # So we could make this a binary 0 and 1 corresponding relationship
    #if the comment is fat shaming its 1 and if its skinny shaming its 0
    #if the Expected Categorization is toxic then its 1 and if its not toxic its 0
    #if the predicted toxicity categorization is toxic then its 1 if its not toxic then its 0
    X['Type of Body Shaming'] = [1 if x == 'Fat Shaming' else 0 for x in df['Type of Body Shaming']]
    X['Expected Categorization'] = [1 if x == 'Toxic' else 0 for x in df['Expected Categorization']]
    X['Predicted Toxicity Categorization'] = [1 if x == 'Toxic' else 0 for x in df['Predicted Toxicity Categorization']]

#If the comment is in group 1 it is a fat shaming comment, and if it is in group 2 it is a skinny shaming comment

    comment_groups = {
        'She is such a fatty she ate a whole box of crumble cookies': 0,
        'He looks fatter than a bloated whale in the water when he is swimming': 0,
        'He would be so much hotter if he was not an ugly obese bitch': 0,
        'She thought she looked good but her fat body looked so huge in that dress': 0,
        'His friends are so fit, but he is the only one that looks like a fat obese ogre': 0,
        'She eats so much during lunch no wonder she is such a fat pig': 0,
        'Those jeans do not look good on her skinny stick legs': 1,
        'How did she win the pageant she looks flat and thin in that outfit': 1,
        'He needs to eat more why is he shaped like a skinny twig': 1,
        'Are you on a diet you have lost so much weight your unattractive body looks so weak': 1,
        'That girl was flatter than a wall I feel bad for her skinny bony body': 1,
        'Look at her nasty lanky legs and arms she looks like a 5 year old': 1
    }

    X['Analyzed Comment'] = [comment_groups.get(str(x).strip()) for x in df['Analyzed Comment']]

    return X

# Example usage:
X_test = featurize(test_df)
X_test.head()
```

```python
# extract indices of examples for the shaming comments

body_shaming_column = test_df["Type of Body Shaming"]

fat_shaming_comments = []
skinny_shaming_comments = []

for i in range(len(body_shaming_column)):
    if body_shaming_column[i] == "Fat Shaming":
        fat_shaming_comments.append(i)
    elif body_shaming_column[i] == "Skinny Shaming":
        skinny_shaming_comments.append(i)

y_actual_fat_shaming = [y_actual[i] for i in fat_shaming_comments]
y_predicted_fat_shaming = [y_predicted[i] for i in fat_shaming_comments]

y_actual_skinny_shaming = [y_actual[i] for i in skinny_shaming_comments]
y_predicted_skinny_shaming = [y_predicted[i] for i in skinny_shaming_comments]

print (len(fat_shaming_comments))
print (len(skinny_shaming_comments))
```

```python
def accuracy(y_actual, y_predicted):
    total_p = 0
    total_n = 0
    TP=0
    TN=0
    for i in range(len(y_predicted)):
        if y_actual[i]==1:
            total_p = total_p+1
            if y_actual[i]==y_predicted[i]:
               TP=TP+1
        if y_actual[i]==0:
            total_n=total_n+1
            if y_actual[i]==y_predicted[i]:
               TN=TN+1
    return(TP/total_p, TN/total_n)

acc_1_fat_shaming, acc_0_fat_shaming = accuracy(y_actual_fat_shaming, y_predicted_fat_shaming)
acc_1_skinny_shaming, acc_0_skinny_shaming = accuracy(y_actual_skinny_shaming, y_predicted_skinny_shaming)

print (f"Fat Shaming 1 accuracy for Fat Shaming Comments = {acc_1_fat_shaming}")
print (f"Fat Shaming 0 accuracy for Fat Shaming Comments = {acc_0_fat_shaming}")
print (f"Skinny Shaming 1 accuracy for Skinny Shaming Comments = {acc_1_skinny_shaming}")
print (f"Skinny Shaming 0 accuracy for Skinny Shaming Comments = {acc_0_skinny_shaming}")
```
